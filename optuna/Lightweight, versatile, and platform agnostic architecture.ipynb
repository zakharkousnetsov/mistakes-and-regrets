```python
import optuna
```

```python
def objective(trial):
    x = trial.suggest_float("x", -10, 10)
    return (x - 2) ** 2
```

```python
%%time
study = optuna.create_study()
study.optimize(objective, n_trials=10000, timeout=12)
```
